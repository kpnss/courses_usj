```python
# Imports
import sklearn as sk
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sp
import seaborn as sb

from sklearn.cluster import DBSCAN
```

```python
# Cargar dataset
countries = pd.read_csv('country_data.csv')
print(countries)
np.shape(countries)[1]
countries.columns
```

```python
# Estandardizar datos
dataframe = countries.iloc[:, 0:10]
dataframe[dataframe.columns[1:10]] = dataframe[dataframe.columns[1:10]].astype(float)

dataframe[dataframe.columns[1:10]] = sk.preprocessing.StandardScaler().fit_transform(dataframe[dataframe.columns[1:10]])

print(np.mean(dataframe.iloc[:,1]))
print(np.std(dataframe.iloc[:,1]))
print(dataframe)
```

```python
epsilons = [0.5, 0.75, 1]
min_points = [2,4,6]

for e in epsilons:
    for p in min_points:

        clustering = DBSCAN(eps=e, min_samples=p).fit(dataframe[dataframe.columns[1:10]])
        labels = clustering.labels_
        print(f"\n DBSCAN epsilon: {e}, min_points: {p}\n", labels)
        wcss = 0
        unique_labels = set(labels)
        for label in unique_labels:
            if label == -1:  # No tiene sentido calcular el centroide para los puntos etiquetados como ruido
                continue
            
            cluster_points = dataframe[dataframe.columns[1:10]][labels == label]
            
            centroid = np.mean(cluster_points, axis=0)
            
            wcss += np.sum((cluster_points - centroid) ** 2)
        anomalies = dataframe[labels == -1]
        print("Anomalías:", len(anomalies))
        
        '''
        for country, cluster in zip(dataframe.iloc[:, 0], labels):
                print(f"{country}: cluster {cluster}")
        '''
        print("Inercia", wcss)
        print("Numero clusteres:",len(unique_labels)-1 ) # -1 para no contar el cluster de ruido
```

Considerando el valor de la inercia como criterio de elección, excluyendos los casos de inercia cero, el mejor algoritmo sería el que utiliza un épsilon de 0.5 y un min_points de 2. En este caso, obtendríamos 158 anomalías (ruido) y solo 3 clústeres. 

```python
epsilons = [1, 1.5, 2, 2.5, 3]
min_points = [2,4,6]

for e in epsilons:
    for p in min_points:

        clustering = DBSCAN(eps=e, min_samples=p).fit(dataframe[dataframe.columns[1:10]])
        labels = clustering.labels_
        print(f"\n DBSCAN epsilon: {e}, min_points: {p}\n", labels)
        wcss = 0
        unique_labels = set(labels)
        for label in unique_labels:
            if label == -1:  # No tiene sentido calcular el centroide para los puntos etiquetados como ruido
                continue
            
            cluster_points = dataframe[dataframe.columns[1:10]][labels == label]
            
            centroid = np.mean(cluster_points, axis=0)
            
            wcss += np.sum((cluster_points - centroid) ** 2)
        anomalies = dataframe[labels == -1]
        print("Anomalías:", len(anomalies))
        
        '''
        for country, cluster in zip(dataframe.iloc[:, 0], labels):
                print(f"{country}: cluster {cluster}")
        '''
        print("Inercia", wcss)
        print("Numero clusteres:",len(unique_labels)-1 ) # -1 para no contar el cluster de ruido
```

Sin considerar la inercia, para elegir los mejores hiperparámetros tuve que realizar una búsqueda manual. Los parámetros min_points y epsilon modifican cómo se agrupan los clústeres. Al probar con varios valores, noté que con un épsilon menor a 1 los clústeres no se unían (lo que provocaba que demasiados países fueran etiquetados como anomalías). Por ello, probé con valores más altos, buscando un equilibrio para evitar la creación de un único "mega clúster".

Los valores óptimos son aquellos que no agrupan todos los puntos en un mismo clúster (0) pero que tampoco generan un exceso de anomalías. Los parámetros que mejor se ajustan a esta descripción son:

    Epsilon 1, min_points 4: Resulta en 85 anomalías y 4 clústeres.

    Epsilon 1.5, min_points 2: Resulta en 23 anomalías y 3 clústeres.

Para elegir el mejor entre ambos, procedo a analizar las etiquetas asignadas a cada país.

```python
clustering = DBSCAN(eps=1, min_samples=4).fit(dataframe[dataframe.columns[1:10]])
labels = clustering.labels_

for country, cluster in zip(dataframe.iloc[:, 0], labels):
    print(f"{country}: cluster {cluster}")
```

```python
clustering = DBSCAN(eps=1.5, min_samples=2).fit(dataframe[dataframe.columns[1:10]])
labels = clustering.labels_

for country, cluster in zip(dataframe.iloc[:, 0], labels):
    print(f"{country}: cluster {cluster}")
```

Entre estos dos, considero que el mejor es epsilon 1, min_points 4, ya que el otro une países que son demasiado diferentes entre sí. Con esta configuración obtenemos un clúster más y un mayor número de anomalías; esto indica que el algoritmo DBSCAN está siendo más selectivo, aunque no logremos agrupar una gran cantidad de países en un solo clúster.

Sin embargo desde ahora usaremos los parametros calculados antes, aquellos que tienen mejor inercia.

```python
# Silhoutte score
from sklearn.metrics import silhouette_score
# Kmeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(dataframe.iloc[:, 1:10])
print(kmeans.inertia_)
clusters = kmeans.fit_predict(dataframe.iloc[:,1:10])

score = silhouette_score(dataframe.iloc[:,1:10], clusters)
print(f"Silhouette Score: {score}")

for country, cluster in zip(dataframe.iloc[:, 0], clusters):
    print(f"{country}: cluster {cluster}")
```

```python
# Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters_hc = hc.fit_predict(dataframe.iloc[:, 1:10])
score = silhouette_score(dataframe.iloc[:,1:10], clusters_hc)

print(f"Silhouette Score: {score}")

for country, cluster in zip(dataframe.iloc[:, 0], clusters_hc):
    print(f"{country}: cluster {cluster}")
```

```python
# DBSCAN 
# Los iperparametros elegidos son los que han dado mejor inercia, como pedido por el enunciado, aunque no sea el mejor resultado
clustering = DBSCAN(eps=0.5, min_samples=2).fit(dataframe[dataframe.columns[1:10]])
labels = clustering.labels_
score = silhouette_score(dataframe.iloc[:,1:10], labels)
print(f"Silhouette Score: {score}")
for country, cluster in zip(dataframe.iloc[:, 0], labels):
    print(f"{country}: cluster {cluster}")
```

Por lo tanto, el mejor modelo según el Silhouette Score es K-Means, ya que obtuvo un valor ligeramente superior al del Clustering Aglomerativo. Por el contrario, el resultado de DBSCAN, al estar más abajo de cero, resulta prácticamente erróneo para este conjunto de datos.

De hecho, según la escala que hemos visto en clase, una puntuación superior a 0,25 indica que 'algunos puntos están mal agrupados o cerca de los límites'; esto significa que, aunque no hemos obtenido una clasificación perfecta, es el mejor resultado que podíamos alcanzar con estos parámetros.

```python
# KMeans analisis
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(dataframe.iloc[:, 1:10])
clusters = kmeans.predict(dataframe.iloc[:,1:10])
# Por cada cluster calculamos 
# child_mort
for i in range(3):
    cluster_data = dataframe.iloc[:, 1:10][clusters == i]
    mean_child_mort = cluster_data['child_mort'].mean()
    print(f"Cluster {i}: Media child_mort = {mean_child_mort}")
# exports
for i in range(3):
    cluster_data = dataframe.iloc[:, 1:10][clusters == i]
    mean_exports = cluster_data['exports'].mean()
    print(f"Cluster {i}: Media exports = {mean_exports}")
# income
for i in range(3):
    cluster_data = dataframe.iloc[:, 1:10][clusters == i]
    mean_income = cluster_data['income'].mean()
    print(f"Cluster {i}: Media income = {mean_income}")
# total_fer
for i in range(3):
    cluster_data = dataframe.iloc[:, 1:10][clusters == i]
    mean_total_fer = cluster_data['total_fer'].mean()
    print(f"Cluster {i}: Media total_fer = {mean_total_fer}")
# mean child_mort
print(f"\nMedia child_mort = {dataframe['child_mort'].mean()}")
# mean exports
print(f"Media exports = {dataframe['exports'].mean()}")
# mean income
print(f"Media income = {dataframe['income'].mean()}")
# mean total_fer
print(f"Media total_fer = {dataframe['total_fer'].mean()}")
```

```python
import plotly.express as px
import pandas as pd

dataframe['cluster'] = clusters.astype(str)
dataframe['cluster_label'] = dataframe['cluster'].map({ # Nombre clusters
    '0': 'Cluster 0',
    '1': 'Cluster 1',
    '2': 'Cluster 2'
})

# Sunburst plot
fig = px.sunburst(
    dataframe,
    path=['cluster_label', 'country'],  
    title='Sunburst Plot - Paesi per Cluster (KMeans)',
    color='cluster_label',
    color_discrete_sequence=px.colors.qualitative.Set2
)
plt.figure(figsize=(50, 40))

fig.update_layout(margin=dict(t=50, l=0, r=0, b=0))
fig.show()
```

A traves de esta imagen se nota como el Cluster 0 es el mas largo que todos. Mas que la mitad del total de los paises partenece a esto. Y che muy pocos paises partenecen al Cluster 1.

3. Cluster 0: Países en via de Desarrollo (el cluster mas largo)

Este grupo se sitúa en un terreno intermedio, aunque con unos indicadores positivos. Sus niveles de mortalidad infantil (-0.39) y fertilidad (-0.40) están por debajo de la media global, lo que indica avances significativos en salud y desarrollo social. Sus ingresos y exportaciones están cerca del promedio, representando probablemente a economías emergentes o países en vías de desarrollo.
