```python
# Imports
import sklearn as sk
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sp
import seaborn as sb

from sklearn.cluster import DBSCAN
```

```python
# Cargar dataset
countries = pd.read_csv('country_data.csv')
print(countries)
np.shape(countries)[1]
countries.columns
```

```python
# Estandardizar datos
dataframe = countries.iloc[:, 0:10]
dataframe[dataframe.columns[1:10]] = dataframe[dataframe.columns[1:10]].astype(float)

dataframe[dataframe.columns[1:10]] = sk.preprocessing.StandardScaler().fit_transform(dataframe[dataframe.columns[1:10]])

print(np.mean(dataframe.iloc[:,1]))
print(np.std(dataframe.iloc[:,1]))
print(dataframe)
```

```python
epsilons = [0.5, 0.75, 1]
min_points = [2,4,6]

for e in epsilons:
    for p in min_points:

        clustering = DBSCAN(eps=e, min_samples=p).fit(dataframe[dataframe.columns[1:10]])
        labels = clustering.labels_
        print(f"epsilon: {e}, min_points: {p}\n", labels)
        wcss = 0
        unique_labels = set(labels)
        for label in unique_labels:
            if label == -1:  # Salta gli outlier
                continue
            # Extraer puntos de este clúster
            cluster_points = dataframe[dataframe.columns[1:10]][labels == label]
            # Calcular el centroide del clúster
            centroid = np.mean(cluster_points, axis=0)
            # Sumar distancias al cuadrado al centroide
            wcss += np.sum((cluster_points - centroid) ** 2)
        for country, cluster in zip(dataframe.iloc[:, 0], labels):
                print(f"{country}: cluster {cluster}")
        print("Inercia", wcss)
        print("Numero clusteres:",len(unique_labels))
```

```python
# Silhoutte score

# Kmeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(dataframe.iloc[:, 1:10])
print(kmeans.inertia_)
clusters = kmeans.fit_predict(dataframe.iloc[:,1:10])
for country, cluster in zip(dataframe.iloc[:, 0], clusters):
    print(f"{country}: cluster {cluster}")
```
