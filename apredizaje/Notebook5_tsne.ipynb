# t-SNE

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
```

## Dataset

Genera un dataset utilizado la función `make_moons` de `sklearn`

```python
# Generar un dataset con clusters bien definidos
n_samples = 1000
noise = 0.1
ds = make_moons(n_samples=n_samples, noise=noise)
```

## Perplejidad

Entrena el modelo t-SNE para diferentes valores de **perplejidad**, y visualiza cómo son los datos proyectados en 2D

```python
perp = [0.5, 1, 2, 5]

# unpack dataset returned by make_moons -> (X, y)
X, y = ds

for i in perp:
    tsne = TSNE(n_components=2, perplexity=i, learning_rate=0.05, max_iter=1000)
    X_tsne = tsne.fit_transform(X)
    plt.figure()
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=10)
    plt.title(f"Perplexity {i}")
    plt.xlabel("t-SNE 1")
    plt.ylabel("t-SNE 2")
    plt.show()
```

## Iteraciones

Entrena el modelo t-SNE para diferentes valores de **número de iteraciones**, y visualiza cómo son los datos proyectados en 2D

```python
perp = [10, 30, 50, 100, 999]

# unpack dataset returned by make_moons -> (X, y)
X, y = ds

for i in perp:
    tsne = TSNE(n_components=2, perplexity=i, learning_rate=0.05, max_iter=1000)
    X_tsne = tsne.fit_transform(X)
    plt.figure()
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=10)
    plt.title(f"Perplexity {i}")
    plt.xlabel("t-SNE 1")
    plt.ylabel("t-SNE 2")
    plt.show()
```
