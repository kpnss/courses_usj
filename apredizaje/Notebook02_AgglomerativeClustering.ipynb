```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

# 1. Cargar el dataset

```python
X = pd.read_csv('./Mall_Customers_agg_clust.csv')
X.drop(columns='CustomerID', inplace=True)
X['Gender'] = X['Gender'].map({'Male': 1, 'Female': 0})
```

# 2. Representa el dendrograma

Elige un criterio de fusión y representa el dendrograma usando `scipy` (https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html)

```python
method = "ward"
Z = linkage(X, method)
```

```python
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
```
