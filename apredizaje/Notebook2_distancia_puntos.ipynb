# Distancia entre puntos en altas dimensiones

¿Qué ocurre cuando generamos un punto aleatorio en un espacio de muchas dimensiones? La respuesta (nuevamente) puede resultar anti-intuitiva: En un espacio de dos o tres dimensiones los puntos cercanos tienen valores similares. Sin embargo, en espacios de alta dimensión, las distancias entre puntos tienden a volverse similares. Esto significa que, incluso si dos puntos parecen estar "cercanos" en términos de coordenadas, su distancia real podría ser grande debido a la cantidad de dimensiones en las que pueden diferir.
