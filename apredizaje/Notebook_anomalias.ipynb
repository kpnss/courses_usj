```python
import pandas as pd
import numpy as np
import sklearn as sk 
import scipy.stats as sp
```

```python
# Cargar datos

coches = pd.read_csv('/home/michele/UNI/zaragoza/apredizaje/car_price_dataset_anomalies.csv')
print(coches)
coches_prices = coches['Price']
print(coches_prices)    
```

```python
# Deteccion de anomalias

# IQR
IQR = sp.iqr(coches_prices)
Q1 = np.percentile(coches_prices, 25)
Q3 = np.percentile(coches_prices, 75)

print(f"IQR: {IQR}")
print(f"Q1: {Q1}, Q3: {Q3}")

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in coches_prices if x < lower_bound or x > upper_bound]
print(f"Anomalias IQR: {outliers}")

# Z-score
mean_price = np.mean(coches_prices)

std_price = np.std(coches_prices)
print(f"media: {mean_price}, desviacion estandar: {std_price}")
z_scores = [(x - mean_price) / std_price for x in coches_prices]
outliers_z = [coches_prices[i] for i in range(len(z_scores)) if abs(z_scores[i]) > 3]
print(f"Anomalias Z-score: {outliers_z}")

# MAD
median_price = np.median(coches_prices)
mad = np.median([abs(x - median_price) for x in coches_prices])
print(f"mediana: {median_price}, MAD: {mad}")
mad_scores = [(x - median_price) / (1.4826 * mad) for x in coches_prices]
outliers_mad = [coches_prices[i] for i in range(len(mad_scores)) if abs(mad_scores[i]) > 3]
print(f"Anomalias MAD: {outliers_mad}")
```
